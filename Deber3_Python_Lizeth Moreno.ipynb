### 2. Lectura de Datos

Accedemos a los datos de forma remota mediante el uso de una URL previamente definida. De manera consistente con la tarea anterior, recurrimos a la función 'pd.read_csv' de pandas para la carga eficiente de datos tabulares.

```python
#leer datos con Pandas
url = 'https://raw.githubusercontent.com/cienciadedatos/datos-de-miercoles/master/datos/2019/2019-04-10/partidos.txt'
datos = pd.read_csv(url, delimiter='\t', skiprows=0, dtype=str, encoding='utf-8')
#Visualizar los primeros registros
print(datos.head())
```

### 3. Creación de gráficos que comparen paises

#### 3.1 Gráfico 1 deberá ser un gráfico en el que los datos de los países comparados aparezcan en el mismo cuadro

Primero graficamos por separado la información de nuestros dos paises:


### Argentina

Se escoge Argentina considerado unicamente del equipo 1. Así la variable equipo_1_final me cuenta el número de goles que metió Argentina a través de los años

```python
pais_1 = 'Argentina'
datos_1 = datos[datos['equipo_1'] == pais_1]
datos_1 = datos_1[['anio', 'equipo_1_final']]
datos_1.head()
```

```python
datos_1['equipo_1_final'] = pd.to_numeric(datos_1['equipo_1_final'], errors='coerce')
datos_1.plot(x='anio', y='equipo_1_final', label='Argentina', kind='line',  figsize=(20, 6))
# Etiquetas y título
plt.title('Goles de Argenina a través del tiempo')
plt.xlabel('Año')
plt.ylabel('Goles')
plt.grid(False)
plt.show()
```

```python
datos_1.plot(x='anio', y='equipo_1_final', label='Argentina' ,kind='bar', figsize=(20, 6))
# Etiquetas y título
plt.title('Goles de Argenina a través del tiempo')
plt.xlabel('Año')
plt.ylabel('Goles')
plt.grid(False)
plt.show()
```

```python
datos_1.plot(x='anio', y='equipo_1_final',label='Argentina', kind='area', figsize=(20, 6))
```

Las gráficas previas representan la evolución del número de goles anotados por Argentina a lo largo del tiempo. De estos gráficos, se puede inferir que en ciertos años, como 1966, 1978 y 1982, entre otros, Argentina no registró ninguna anotación en sus partidos. En contraste, se observa que los años 1930, 1978, 1998 y 2006 se destacan como períodos excepcionales en los que Argentina tuvo un desempeño notable, evidenciado por una cantidad significativa de goles anotados.

### Francia

```python
datos_2['equipo_2_final'] = pd.to_numeric(datos_2['equipo_2_final'], errors='coerce')
datos_2.plot(x='anio', y='equipo_2_final', kind='line',label='Francia', figsize=(20, 6))
# Etiquetas y título
plt.title('Goles de Francia a través del tiempo')
plt.xlabel('Año')
plt.ylabel('Goles')
plt.grid(False)
plt.show()
```

```python
datos_2.plot(x='anio', y='equipo_2_final',label='Francia', kind='bar', figsize=(20, 6))
# Etiquetas y título
plt.title('Goles de Francia a través del tiempo')
plt.xlabel('Año')
plt.ylabel('Goles')
plt.grid(False)
plt.show()
```


Con notoria disparidad, en el caso de Francia se observa una marcada consistencia en la cantidad de goles anotados, mostrando una menor variabilidad a lo largo del tiempo. Destaca que, en particular, dos años sobresalen como los mejores para Francia: 1958 y 2014, durante los cuales lograron anotar 6 y 5 goles respectivamente, constituyendo sus años más destacados en términos de desempeño goleador.

### Combinados

Combinamos las bases de datos seleccionando a Argentina como el equipo 1 y a Francia como el equipo 2. Esta concatenación nos brindará la oportunidad de realizar una comparación directa de los goles anotados entre estos dos países.

```python
# Realiza una fusión (merge) basada en la columna 'anio'
datos_combinados = pd.merge(datos_1, datos_2, on='anio', how='inner')
datos_combinados
```

```python
# Convertimos a datos numéricos
datos_combinados['equipo_1_final'] = pd.to_numeric(datos_combinados['equipo_1_final'], errors='coerce')
datos_combinados['equipo_2_final'] = pd.to_numeric(datos_combinados['equipo_2_final'], errors='coerce')

# Graficamos
datos_combinados.plot(x='anio', y=['equipo_1_final', 'equipo_2_final'], linewidth=2,  markersize=0.05, figsize=(20, 6))

# Etiquetamos y agregamos títulos
plt.title('Goles anotados de Argentina vs Francia')
plt.xlabel('Año')
plt.ylabel('Resultado final')
plt.grid(False)
plt.show()
```

La gráfica previa presenta la comparación de goles entre Argentina y Francia. En ella, se destaca que Francia exhibe una mayor constancia en sus anotaciones en comparación con Argentina. No obstante, ambos países muestran, al menos, más de una anotación en el período comprendido entre 1982 y 2002.

#### 3.2 Gráfico 2 deberá ser un gráfico en el que los datos de cada país aparezca en un subgráfico diferente. Es decir, una figura tenga al menos dos recuadros (uno por cada país en la comparación).



```python
fig, axs = plt.subplots(1, 2, figsize=(25, 4))

# Graficar la primera subgráfica 
axs[0].plot(datos_1['anio'], datos_1['equipo_1_final'], label=' Argentina',  linewidth=1)
axs[0].set_xlabel('Año')
axs[0].set_ylabel('Goles Argentina')
axs[0].grid(False)
axs[0].legend()

# Graficar la segunda subgráfica 
axs[1].plot(datos_2['anio'], datos_2['equipo_2_final'], label='Francia',  linewidth=1)
axs[1].set_xlabel('Año')
axs[1].set_ylabel('Goles Francia')
axs[1].grid(False)
axs[1].legend()

# Ajustes extra
plt.suptitle('Comparación de Goles entre Argentina y Francia a través del tiempo', fontsize=16)
plt.show()
```

La gráfica anterior nos proporciona una visualización comparativa de los goles anotados por Argentina y Francia a lo largo del tiempo, con cada país representado en una subgráfica separada.
